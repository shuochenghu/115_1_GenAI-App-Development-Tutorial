## 1. 本週定位：從 Colab 轉到本機 App 專案

第 7 週開始導入 VS Code、本機 Python、Streamlit 與 GitHub。原因不是 Notebook 不重要，而是期中與期末專題需要交付「可以操作的 Web App」。

### Streamlit 是什麼？

Streamlit 是一個 Python Web App 框架。它讓我們用熟悉的 Python 寫互動式網頁，不需要先學 HTML、CSS 或 JavaScript。對本課程來說，Streamlit 的角色是把前幾週寫在 notebook 裡的 AI 功能，包成一般使用者能在瀏覽器操作的工具。

最小的 Streamlit App 只有幾行：

```python
import streamlit as st

st.title("AI 課程助理")
question = st.text_input("請輸入問題")

if question:
    st.write("你輸入的是：", question)
```

這段程式存成 `app.py` 後，在 terminal 執行：

```bash
streamlit run app.py
```

瀏覽器會開啟本機網頁，使用者就能輸入文字、按按鈕、看模型回覆，而不是只看 notebook cell 下方的輸出。

### 為什麼本課程需要 Streamlit？

前幾週我們已經會呼叫 OpenAI API、設計 prompt、做 streaming、處理 structured output；但這些能力如果只停在 notebook 裡，通常只有開發者自己能操作。Streamlit 補上的是「應用程式介面」這一層。

| Notebook / Colab | Streamlit Web App |
|---|---|
| 適合學概念、逐格執行、觀察中間結果 | 適合做成可操作、可展示、可部署的工具 |
| 使用者需要看懂程式 cell | 使用者只需要操作輸入框、按鈕、上傳檔案 |
| 對話紀錄常暫存在變數或 cell 執行順序中 | 對話紀錄放在 `st.session_state` 中管理 |
| API key 常用 Colab Secrets | 本機用 `.env`，部署後用 `st.secrets` |
| 成果多半是教學或實驗 | 成果可以變成期中/期末專題原型 |

### Streamlit 的核心心智模型

Streamlit 看起來像一般網頁，但寫法仍然是 Python script。每次使用者輸入文字、按按鈕、切換選項時，Streamlit 會把 `app.py` 從上到下重新執行一次，並依照目前狀態重新畫出畫面。

因此本週會特別重視三件事：

1. **畫面元件**：用 `st.title()`、`st.write()`、`st.chat_input()`、`st.form()`、`st.file_uploader()` 建立操作介面。
2. **狀態管理**：用 `st.session_state` 保存聊天紀錄、設定值與重置狀態，避免每次重跑就忘記資料。
3. **安全與部署**：用 `.env`、`.gitignore`、`st.secrets` 管理 API key，讓 App 可以安全地放到 GitHub 或部署平台。
