## 3. API Key 管理：本機用 `.env`，部署用 Streamlit Secrets

本機開發時使用 `.env`：

```env
OPENAI_API_KEY=sk-proj-你的key
OPENAI_MODEL=gpt-5.4-mini
```

部署到 Streamlit Community Cloud 時，不上傳 `.env`，改在 App 的 Secrets 設定：

```toml
OPENAI_API_KEY = "sk-proj-你的key"
OPENAI_MODEL = "gpt-5.4-mini"
```

重點：API key 不寫死在 `app.py`，也不推上 GitHub。


````python
import os
from dotenv import load_dotenv


def get_secret(name, default=None):
    """先讀 Streamlit Secrets；本機開發時再改讀 .env。"""
    try:
        import streamlit as st
        if name in st.secrets:
            return st.secrets[name]
    except Exception:
        # 本機或 notebook 環境可能沒有 st.secrets，忽略後改讀 .env。
        pass

    load_dotenv()
    return os.getenv(name, default)


api_key = get_secret("OPENAI_API_KEY")
model = get_secret("OPENAI_MODEL", "gpt-5.4-mini")
print("模型：", model)
print("API key 是否已設定：", bool(api_key))
````

## 4. OpenAI Responses API：App 端最小 helper

本週延續第 3–6 週主線，使用 `client.responses.create()`。Streamlit App 中建議先把 API 呼叫包成函式，避免 UI 程式碼和 API 邏輯混在一起。


````python
from openai import OpenAI


def create_client():
    """建立 OpenAI client，並在缺少 API key 時給出明確錯誤。"""
    api_key = get_secret("OPENAI_API_KEY")
    if not api_key:
        raise RuntimeError("找不到 OPENAI_API_KEY，請設定 .env 或 Streamlit Secrets。")
    return OpenAI(api_key=api_key)


def ask_ai(user_input, system_prompt="你是有幫助的 AI 助理。"):
    """非串流版本：適合摘要、分類等一次性任務。"""
    client = create_client()
    model = get_secret("OPENAI_MODEL", "gpt-5.4-mini")
    response = client.responses.create(
        model=model,
        instructions=system_prompt,
        input=user_input,
    )
    return response.output_text
````

## 6. Session State：保存聊天紀錄

Streamlit 每次互動都會重新執行 `app.py`。如果直接用普通 list 存訊息，按一次按鈕後資料就可能消失。`st.session_state` 是本週最重要的 App 狀態管理工具。


````python
import streamlit as st


def init_messages():
    """初始化聊天紀錄；Streamlit rerun 後仍會保留 session_state。"""
    if "messages" not in st.session_state:
        st.session_state.messages = []


def add_message(role, content):
    """把一則訊息加入聊天紀錄，role 通常是 user 或 assistant。"""
    st.session_state.messages.append({"role": role, "content": content})


def reset_messages():
    """清空聊天紀錄，搭配 st.rerun() 讓畫面立即更新。"""
    st.session_state.messages = []
````

---
## 7. Chat UI：`st.chat_message` 與 `st.chat_input`

聊天介面的基本流程：

1. 初始化 `st.session_state.messages`
2. 逐筆畫出既有訊息
3. 等待使用者用 `st.chat_input()` 輸入
4. 呼叫模型
5. 把 user 與 assistant 訊息都存回 session state


````python
# Chat UI 片段，請放入 app.py
init_messages()

# 每次 rerun 都先重畫歷史訊息。
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.write(message["content"])

prompt = st.chat_input("請輸入問題")
if prompt:
    add_message("user", prompt)
    with st.chat_message("user"):
        st.write(prompt)

    with st.chat_message("assistant"):
        answer = ask_ai(prompt)
        st.write(answer)
    add_message("assistant", answer)
````

---
## 8. Streaming：讓回覆逐步顯示

第 5 週已在 Notebook 中看過 `response.output_text.delta`。在 Streamlit 中，我們可以寫一個 generator，逐段 `yield` delta，再交給 `st.write_stream()` 顯示。


````python
def stream_ai(user_input, system_prompt="你是有幫助的 AI 助理。"):
    """串流版本：逐段 yield 文字，交給 st.write_stream() 顯示。"""
    client = create_client()
    model = get_secret("OPENAI_MODEL", "gpt-5.4-mini")
    stream = client.responses.create(
        model=model,
        instructions=system_prompt,
        input=user_input,
        stream=True,
    )
    for event in stream:
        # Responses API streaming 會送出多種事件；這裡只取文字 delta。
        if getattr(event, "type", None) == "response.output_text.delta":
            yield event.delta
````

````python
# 在 chat_input 區塊中使用 streaming
if prompt:
    add_message("user", prompt)
    with st.chat_message("user"):
        st.write(prompt)

    with st.chat_message("assistant"):
        # st.write_stream() 會顯示串流文字，並回傳完整回答字串。
        full_answer = st.write_stream(stream_ai(prompt, system_prompt))
    add_message("assistant", full_answer)
````

## 9. Sidebar：模型、角色與重置按鈕

Sidebar 適合放不需要佔用主畫面的設定，例如角色、回答風格、是否串流、重置聊天紀錄。不要把 API key 顯示在畫面上。


````python
# Sidebar 放全域設定，不佔用主要互動畫面。
with st.sidebar:
    st.header("設定")
    system_prompt = st.text_area("System prompt", value="你是有幫助的 AI 助理。")
    use_streaming = st.toggle("使用串流回覆", value=True)
    if st.button("清空聊天紀錄"):
        reset_messages()
        st.rerun()
````

## 10. 表單：把 Prompt Engineering 做成工具

聊天介面適合自由問答；表單適合固定任務，例如摘要、改寫、分類與資料抽取。表單能讓使用者輸入欄位更明確，也方便你控制 prompt。


````python
# 表單適合固定任務：使用者填完欄位後再一次送出。
with st.form("summary_form"):
    source_text = st.text_area("貼上要摘要的文字", height=160)
    style = st.selectbox("摘要風格", ["條列重點", "一段式摘要", "給主管看的摘要"])
    submitted = st.form_submit_button("產生摘要")

if submitted and source_text:
    prompt = f"請將以下內容整理成{style}：\n\n{source_text}"
    summary = ask_ai(prompt, system_prompt="你是嚴謹的摘要助理。")
    st.subheader("摘要結果")
    st.write(summary)
````

## 11. 檔案上傳：期中專題常見入口

本週先處理文字檔 `.txt` / `.md`，讓學生理解檔案上傳流程。PDF、Word、CSV 會在第 9 週正式處理。


````python
# 第 7 週先處理純文字檔；PDF / Word / CSV 會在後續文件處理週延伸。
uploaded_file = st.file_uploader("上傳 .txt 或 .md 檔案", type=["txt", "md"])
if uploaded_file is not None:
    text = uploaded_file.read().decode("utf-8")
    st.text_area("檔案內容預覽", text[:2000], height=200)
    if st.button("摘要這份檔案"):
        # 控制輸入長度，避免單次測試成本過高。
        prompt = "請摘要以下文字，列出 5 個重點：\n\n" + text[:12000]
        st.write(ask_ai(prompt, system_prompt="你是文件摘要助理。"))
````
